 ## Sentiment Analysis

```python
lm=WordNetLemmatizer()
```

```python
def clean(text):
    text = str(text).lower()
    text = re.sub('[^a-z]',' ',str(text)) # whatever your are removing other than a and z replace with the space in the str(text)
    text = text.split() #split text into single single words
    text = [lm.lemmatize(j) for j in text if j not in set(stopwords.words('english'))] 
    text = ' '.join(str(k) for k in text)
    return text
```

```python
# function to get 
# polarity (float value that lies btween[-1:1] when 1 means positive sentiment and -1 means negative sentiment)
# subjectivity (if the sentence is subjective or objective----subjective means sentence refer to personal opion and not a fact)
def getpolarity(text):
    return TextBlob(text).sentiment.polarity

def getsubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
```

```python
def getAnalysis(score):
    if score<0 :
        return 'negative'
    elif score ==0:
        return 'neutral'
    else:
        return 'positive'
```

## DONALD TRUMP TWEET SENTIMENT ANALYSIS

```python
trump_tweets = data[data['candidate']=='Trump']
trump_tweets = trump_tweets.loc[trump_tweets.country == 'US']
trump_tweets= trump_tweets[['tweet']]
trump_tweets
```

```python
nltk.download('omw-1.4')
```

```python
trump_tweets['cleantext'] = trump_tweets['tweet']
```

```python
trump_tweets['cleantext'].apply(clean)
trump_tweets
```

```python
trump_tweets['getsubjectivity'] = trump_tweets['cleantext'].apply(getsubjectivity)
```

```python
trump_tweets['polrity'] = trump_tweets['cleantext'].apply(getpolarity)
```

```python
trump_tweets['analysis'] = trump_tweets['polrity'].apply(getAnalysis)
```

```python
trump_tweets.head()
```

```python
plt.figure(figsize=(10,5))
(trump_tweets.analysis.value_counts(normalize=True)*100).plot.bar()
plt.ylabel("%age of tweets")
```

```python
from wordcloud import WordCloud, STOPWORDS
def word_cloud(wd_list):
    stopwords = set(STOPWORDS)
    all_words = ' '.join([text for text in wd_list])
    wordcloud = WordCloud(background_color = 'white', stopwords = stopwords, width = 1600, height = 800, max_words = 100, max_font_size = 200, colormap = 'jet').generate(all_words)
    plt.figure(figsize=(12,10))
    plt.imshow(wordcloud)
word_cloud(trump_tweets['tweet'][:5000])
```

```python
# import libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
fd = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Types-Of-Cross-Validation/main/cancer_dataset.csv')
df = fd.copy()
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,2:] # all rows from the second column
y = df.iloc[:,1] # first column all the rows
```

```python
# KFOLD
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, cross_val_score
k = KFold(5)
model = LogisticRegression()
```

```python
np.mean(cross_val_score(model,x,y, cv = k))
```

```python
model = LogisticRegression()
arr = [2,3,4,5,6]
for i in arr:
    k = KFold(i)
    result = cross_val_score(model,x,y,cv=k)
    print(f'when k={i}',np.mean(result))
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
k = StratifiedKFold(n_splits=5)
result = cross_val_score(model,x,y,cv=k)
np.mean(result)
```

```python
## LOO(Leave One Out) CV
from sklearn.model_selection import LeaveOneOut
lc = LeaveOneOut()
result = cross_val_score(model,x,y,cv=lc)
```

```python
np.mean(result)
```

# HYPER-PARAMETER TUNING

```python
from sklearn import svm, datasets
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df
```

```python
df['target'] = df['target'].apply(lambda x : iris.target_names[x])
df.head()
```

```python
# Tuning
kernels = ['rbf','linear']
C = [1,10,30,20]
avg_scores={}
for kval in kernels:
    for cval in C:
        cv_score = cross_val_score(svm.SVC(kernel= kval, C= cval,gamma='auto'),iris.data,iris.target, cv=5)
        avg_scores[kval + '_' + str(cval)] = np.mean(cv_score)
```

```python
avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'),{'C':[1,5,7,10],'kernel':['rbf','linear','sigmoid','poly'],
                                         'max_iter':[1,5,8,10,20]},cv=10)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
pd.DataFrame(clf.cv_results_)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
clf = RandomizedSearchCV(svm.SVC(gamma='auto'),{'C':[1,5,7,10],'kernel':['rbf','linear','sigmoid','poly'],
                                         'max_iter':[1,5,8,10,20]},cv=10,n_iter=5)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
pd.DataFrame(clf.cv_results_)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_params = {
    'svm': {'model': svm.SVC(gamma='auto'),'params':{'C':[1,10,15,20,25,30,50,100],'kernel':['rbf','linear']}},
    'randomforest' : {'model': RandomForestClassifier(),'params':{'n_estimators':[1,5,10,20,50]}},
    'LogisticRegression':{'model': LogisticRegression(),'params':{'C':[1,10,15,20,25,30,50,100]}}
    
}
```

```python
scores = []
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'],cv=10)
    clf.fit(iris.data,iris.target)
    scores.append({
        'model':model_name,
        'best_score': clf.best_score_,
        'best_params':clf.best_params_
        
    })
dfinal = pd.DataFrame(scores,columns=['model','best_score','best_params'])
dfinal
```
